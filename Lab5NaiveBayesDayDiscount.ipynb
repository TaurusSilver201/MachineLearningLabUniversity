```python
import pandas as pd

# Load Naive_Bayes_Dataset
nb_data = pd.read_excel('Naive_Bayes_Dataset.xlsx')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Encode categorical variables
for col in ['Day', 'Discount', 'Free_Delivery', 'Purchase']:
    nb_data[col] = label_encoder.fit_transform(nb_data[col])
```

```python
# Prepare features and target
X = nb_data[['Day', 'Discount', 'Free_Delivery']]
y = nb_data['Purchase']
```

```python
# Train Naive Bayes model
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(X, y)
```

```python
from sklearn.preprocessing import LabelEncoder
# Make predictions

# Create a new LabelEncoder instance for encoding 'Yes' and 'No' for prediction
prediction_encoder = LabelEncoder()
# Fit this encoder specifically on the values expected for prediction
# Assuming 'No' and 'Yes' are the only possibilities for Discount and Free_Delivery
prediction_encoder.fit(['No', 'Yes'])


try:
    # Use the prediction_encoder to transform 'Yes' and 'No'
    encoded_discount = prediction_encoder.transform(['Yes'])[0]
    encoded_free_delivery = prediction_encoder.transform(['No'])[0]

except ValueError as e:
    print(f"Error during encoding for prediction: {e}")
    print("Please ensure 'Yes' and 'No' are the only possible values for Discount and Free_Delivery, or handle other values appropriately.")
    encoded_discount = None
    encoded_free_delivery = None


if encoded_discount is not None and encoded_free_delivery is not None:
    # The 'Day' column was encoded in a previous step, so use its encoded value
    # Assuming 'Day' was encoded such that 1 corresponds to a valid category from the training data
    sample_data = pd.DataFrame({'Day': [1],
                                'Discount': [encoded_discount],
                                'Free_Delivery': [encoded_free_delivery]})

    print("Sample data after encoding for prediction:")
    display(sample_data)

    # Use the nb_model trained on the original encoded data
    prediction = nb_model.predict(sample_data)

    # Use the prediction_encoder to inverse transform the prediction (0 or 1 back to 'No' or 'Yes')
    purchase_prediction = prediction_encoder.inverse_transform(prediction)
    print(f'Purchase Prediction: {purchase_prediction}')
else:
    print("Cannot make prediction due to encoding error.")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Setup LabelEncoder for Yes/No type features
yn_encoder = LabelEncoder()
yn_encoder.fit(['No', 'Yes'])  # maps: 'No'->0, 'Yes'->1

def make_prediction(nb_model, day_value, discount, free_delivery):

    try:
        encoded_discount = yn_encoder.transform([discount])[0]
        encoded_free_delivery = yn_encoder.transform([free_delivery])[0]

        sample_data = pd.DataFrame({
            'Day': [day_value],
            'Discount': [encoded_discount],
            'Free_Delivery': [encoded_free_delivery]
        })

        print("Sample data after encoding:")
        display(sample_data)

        prediction = nb_model.predict(sample_data)
        purchase_prediction = yn_encoder.inverse_transform(prediction)

        print(f"Purchase Prediction: {purchase_prediction[0]}")

    except ValueError as e:
        print(f"Encoding error: {e}")
        print("Make sure inputs are only 'Yes' or 'No' for Discount and Free_Delivery.")

# ---------------- Example usage ----------------
# Just change these inputs to test new cases:
make_prediction(
    nb_model=nb_model,
    day_value=1,            # Use correct encoded day value from training
    discount='Yes',
    free_delivery='No'
)
```
