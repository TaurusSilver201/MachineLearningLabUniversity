```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# Step 1: Load the dataset
df = pd.read_csv("encoding.csv")
```

```python
# Step 2: Apply Label Encoding (Converting categorical columns into numeric‚ê£labels)
label_enc = LabelEncoder()
df['Color_LabelEncoded'] = label_enc.fit_transform(df['Color'])
df['Size_LabelEncoded'] = label_enc.fit_transform(df['Size'])
```

```python
df
```

```python
# Step 3: Apply One-Hot Encoding (Creating binary columns)
df_onehot = pd.get_dummies(df, columns=['Color', 'Size'], prefix=['Color','Size'])
```

```python
df_onehot
```

```python
# Convert boolean values to integers
df_onehot = df_onehot.astype(int)
```
