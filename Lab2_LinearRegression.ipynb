```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_csv('dhaka homeprices.csv', sep = ';')
```

```python
# Split the 'area,price' column into two separate columns 'area' and 'price'
df[['area', 'price']] = df['area,price'].str.split(',', expand=True)

# Convert the 'area' and 'price' columns to numeric types
df['area'] = pd.to_numeric(df['area'])
df['price'] = pd.to_numeric(df['price'])

plt.xlabel('Area in Square Fit')
plt.ylabel('Price in Taka')
plt.scatter(df['area'], df['price'])
plt.scatter(df['area'], df['price'],color='red', marker='+')
plt.title('Homeprices in Dhaka city')
plt.plot()
```
