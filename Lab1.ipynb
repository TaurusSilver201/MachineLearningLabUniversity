```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('bank.csv', sep = ';')
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
def replace_marital(val):
  if val == "single":
    return 0
  else:
    return 1

df["marital"] = df["marital"].apply (replace_marital,1)
```

```python
df.head(10)
```

```python
df["housing"] = df["housing"].map({
    "no": 0,
    "yes": 1
}.get)
```

```python
df.head(10)
```

```python
df["loan"] = df["loan"].replace({
    "no": 0,
    "yes": 1
})
```

```python
df.head(10)
```

```python
df["job"].unique()
```

```python
df["job"].replace({
    'unknown': np.nan,
    'management': 0,
    'technician': 1,
    'entrepreneur': 2,
    'blue-collar': 3,
    'retired': 4,
    'admin\.': 5,
    'services': 6,
    'self-employed': 7,
    'unemployed': 8,
    'housemaid': 9,
    'student': 10
    },inplace=True)
```

```python
df.head(10)
```

```python
df["education"].unique()
```

```python
df["education"].replace({
    'unknown': np.nan,
    'primary': 0,
    'secondary': 1,
    'tertiary': 2
    },inplace=True)
```

```python
df.head(10)
```

```python
df["contact"].unique()
```

```python
df["contact"].replace({
    'unknown': np.nan,
    'cellular': 0,
    'telephone': 1,
    },inplace=True)
```

```python
df.head(10)
```

```python
df["balance"].min()
```

```python
df["balance"].max()
```

```python
df["balance"] = df["balance"].apply(lambda x: (x - df["balance"].min()) / (df["balance"].max() - df["balance"].min()))
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df["balance"] = scaler.fit_transform(df[["balance"]])
```
